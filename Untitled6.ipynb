```python
import cv2
import smtplib
from email.message import EmailMessage

def send_email(to_email, subject, message):
  """Sends an email with the specified recipient, subject, and message."""

  msg = EmailMessage()
  msg['Subject'] = subject
  msg['From'] = 'yasirmoidutty@gmail.com'
  msg['To'] = to_email
  msg.set_content(message)

  server = smtplib.SMTP('smtp.gmail.com', 587)
  server.starttls()
  server.login('yasirmoidutty@gmail.com', 'qmqcqaowoqwjfmry')
  server.send_message(msg)
  server.quit()

def detect_face(image):
  """Detects faces in the image and returns a list of face bounding boxes."""

  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
  faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
  return faces

def main():
  """Detects faces in the image and sends an email if a face is found."""

  image = cv2.imread('/content/vaisak.jpg')
  faces = detect_face(image)
  if faces.any():
    to_email = 'prj165@iitpkd.ac.in'
    subject = 'Face detected!'
    message = 'A face was detected in the image.'
    send_email(to_email, subject, message)

if __name__ == '__main__':
  main()
```
